# Multiple Linear Regression
- Multiple Linear Regression is an extension of simple Linear regression as it takes more than one predictor variable to predict the response variable.
- i.e. you you have more than one indipendant variable and only one dependent variable at that time we will use Multilinear Regression

Multi Linear Regression Equation:
- Y = m1x1 + m2x2 + m3x3 + ... + C

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\Jahanvi\Data Science\ML\Multiple_Linear_Regression.csv")
```

```python
X = dataset.iloc[:,:-1]
Y = dataset["Salary"]
```

```python
dataset.shape
```

```python
# We can use Multiple linear regression only if our data is 2 dimensional
dataset.ndim
```

```python
# We don't have any seprate data that's why we split this data into train and test
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# Y = m1x1 + m2x2 + c
```

```python
# This will give value of m1 and m2
lr.coef_
```

```python
# This will give value of c
lr.intercept_
```
