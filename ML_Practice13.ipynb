# Logistic Regression(Binary Classification)(Multiple Input)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\Jahanvi\Data Science\ML\placement_classification.csv")
dataset.head(3)
```

- First off all we have to check our graph is linearly seprable or not

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
plt.legend(loc=1)
plt.show()
```

```python
X = dataset.iloc[:,:-1]
Y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)*100
```

```python
lr.predict([[5.9,106]]) #Does not placed
```
