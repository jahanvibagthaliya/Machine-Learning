### How to drop or delete null value from data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Here Credit_History column have maximum null value so we remove that column
# "drop" method used to delete column who contains null values
# Try to delete those column who have maximum null data
# "dataset.drop(columns = ["Credit_History"])" this comand make new dataset
# "dataset.drop(columns = ["Credit_History"], inplace = True)" this command replace data in old dataset

dataset.drop(columns = ["Credit_History"], inplace = True)
```

```python
# remove rows and columns(Who contains null values) are not best everytime sometime we lost important information
# Inplace of remove we fill data over there, this technique is best for machine learning
# How to fill categorical data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Vaishnav\Downloads\loan_data_set.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
# Now we try to fill categorical data who contains null value
# axis=0 means row wise and axis=1 means column wise
# By default it takes axis=0 ex:"dataset.fillna(method="bfill")"
# forward filling - ffill() ex: "dataset.fillna(method="ffill", axis = 1)"  
# Backward filling - bfill()  ex: "dataset.fillna(method="bfill", axis = 1)"
# But ffill and bfill is not correct way to fill null values, sometimes it fills object value to numeric values and numeric values to object values

d1 = dataset.fillna(method="bfill", axis = 1)
d1.head(5)
```

```python
d2 = dataset.fillna(method="ffill", axis =1)
d2.head(5)
```

```python
dataset
```

```python
dataset.info()
```

```python
# Inplace of ffill and bfill we will use mode
# mode() gives maximum repeated value in specific column
# Here for "Gender" column mode value is Male.

dataset["Gender"].mode()
```

```python
dataset["Gender"].mode()[0]
```

```python
# How to fill data in perticular column

dataset["Gender"].fillna(dataset["Gender"].mode()[0], inplace=True)
```

```python
#How to find object data
dataset.select_dtypes(include ="object").isnull().sum()
```

```python
# How to fill data for multiple column

for i in dataset.select_dtypes(include ="object").columns:
    print(i)
```

```python
for i in dataset.select_dtypes(include ="object").columns:
    dataset[i]=dataset[i].fillna(dataset[i].mode()[0])
   
    
```

```python
#we remove all the null value from all the object type and sat it's mode value
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
# How to remove null value from float Dtype and replace it with it's mode value
dataset.select_dtypes(include="float64").isnull().sum()
```

```python
for i in dataset.select_dtypes(include="float64").columns:
    dataset[i]=dataset[i].fillna(dataset[i].mode()[0])
```

```python
#we remove all the null value from all the "float64" type and sat it's mode value
dataset.isnull().sum()
```

### One Hot Encoding & Dummy Variable

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\Vaishnav\Downloads\loan_data_set.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
# Encoding perform on those data who contanis less amount of repeated data
# Here in "Gender" and "Married" column contain only 2 to 3 repeated data

# First of all we fill mode value inplace of null value in "Gender" and "Married" column

#dataset["Married"].fillna(dataset["Married"].mode()[0], inplace=True)
#dataset["Gender"].fillna(dataset["Gender"].mode()[0], inplace=True)
#Above two statement is valid but it gives an error

dataset["Married"]=dataset["Married"].fillna(dataset["Married"].mode()[0])
dataset["Gender"]=dataset["Gender"].fillna(dataset["Gender"].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
#Use First technique using get_dummies() method
en_data = dataset[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
# Above data is in boolean form but we want data in binary form means 0 and 1
# So we use sklearn library
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar, columns =["Gender_Female","Gender_Male","Married_No","Married_Yes"])
```

```python
#Here drop="first" remove first column for every partition column for "Gender_Female" and "Gender_Male" it gives only "Gender_Male" and for "Married_No" and "Married_Yes" it gives only "Married_Yes"
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```

### Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","xxl","m","s","l","xl","xxl","m","s","xl"]})
df.head(3)
```

```python
ord_data = [["s","m","l","xl","xxl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
```

```python
df["Size_en"]=oe.transform(df[["Size"]])
df
```

```python
# While we use map we take any values in any order, here we take 0,1,2,3,4 but we also take it as 7,5,10,99,1
ord_data1 = {"s":0, "m":1, "l":2, "xl":3, "xxl":4}
```

```python
df["Size"].map(ord_data1)
```

```python
df["Size_en_map"]=df["Size"].map(ord_data1)
```

```python
df
```
