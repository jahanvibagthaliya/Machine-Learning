# Polynomial Regression
- Polynomial regression is a regression algorithm that models the relationship between a dependent(Y) and independent variable (X) as n^th degree polynomial.
- Y = B0 +B1X + B2X^2 +B3X^3 + ... + BnX^n

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\Jahanvi\Data Science\ML\polynomial_level_salary.csv")
```

```python
dataset.head(3)
```

```python
plt.scatter(dataset['Level'],dataset['Salary'])
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
dataset.columns
```

```python
dataset.corr()
```

```python
X = dataset[["Level"]]   # Here data always be multidimensional
Y = dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)   # here we can take any degree like 2, 3, 4 etc
pf.fit(X)
X = pf.transform(X)  # Now data is transform into polynomial form
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# Y = m0 +m1X + m2X^2 + C
# Y = 0 + 1657.41487375*X + 186.44305343*X^2 + 4634.5304459095205
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_test,y_test)*100
```

```python
prd = lr.predict(X)
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.plot(dataset["Level"],prd,c="red")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(["org","prd"])
plt.show()
```
