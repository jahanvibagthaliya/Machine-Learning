# Feature Scaling (Standardization)

```python
# Feature Scaling are two types
# 1. Standardization
# 2. Normalization
```

```python
# Defination of standardization in book
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Vaishnav\Downloads\loan_data_set.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# we apply feature scaling technique for "ApplicantIncome" and "CoapplicantIncome"
# For that first we check there is any null value present in "ApplicantIncome" and "CoapplicantIncome"
# If null value present than we remove that null value first
# Like: dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(), inplace=True)

sns.distplot(dataset["ApplicantIncome"])
plt.show()
 
```

```python
dataset.describe()
```

```python
# sklearn provide feature, with the help of that feature we perform automatic scaling
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
# while scaling your data first fit your data and then transform your data
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
dataset.head()
```

```python
dataset.describe()
```

```python
# In above data mean is 0 and variance is 1
# Now we check after scaling nature of our data will be change or not

sns.distplot(dataset["ApplicantIncome_ss"])
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("After")
sns.distplot(dataset["ApplicantIncome_ss"])

plt.subplot(1,2,2)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])

plt.show()
```

```python
# It means after scaling your data will not change just magnitude will be change
```

# Feature Scaling (Normalization)

```python
# Min, max technique it also called Normalization Techniques
# When you use Normalization your data will be reduse according to min or max values
# Here we perform feature scaling for "CoapplicantIncome" column

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
dataset1 = pd.read_csv(r"C:\Users\Vaishnav\Downloads\loan_data_set.csv")
dataset1.head()
```

```python
dataset1.isnull().sum()
```

```python
# In above "CoapplicantIncome" there is no null value present
dataset1.describe()
```

```python
sns.distplot(dataset1["CoapplicantIncome"])
plt.show()
```

```python
# It's look like a normal distribution 
# In Normalizatio, if we perform scaling it will not remove outlier but just magnitude of data will be change

from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset1[["CoapplicantIncome"]])
```

```python
ms.transform(dataset1[["CoapplicantIncome"]])
```

```python
dataset1["CoapplicantIncome.min"]=ms.transform(dataset1[["CoapplicantIncome"]])
dataset1.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset1["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset1["CoapplicantIncome.min"])

plt.show()
```
